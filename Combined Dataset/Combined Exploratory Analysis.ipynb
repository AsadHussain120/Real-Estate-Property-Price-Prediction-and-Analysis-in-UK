```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('combined.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Filtering for Medway
medway_df = data[data['County Name'] == 'Medway']

# Converting the 'Date' column to datetime format
medway_df['Date'] = pd.to_datetime(medway_df['Date'])

# Sorting the dataframe by Date
medway_df = medway_df.sort_values(by='Date')

# Line plot for Average Price over time
plt.figure(figsize=(12, 6))
plt.plot(medway_df['Date'], medway_df['Average price All property types'], marker='o', linestyle='-')
plt.title('Average Price for All Property Types in Medway Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Line plot for Sales Volume over time
plt.figure(figsize=(12, 6))
plt.plot(medway_df['Date'], medway_df['Sales volume'], marker='o', linestyle='-', color='orange')
plt.title('Sales Volume in Medway Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Bar plot for Average Price of Different Property Types
property_types = ['Average price Detached houses', 'Semi-Detached House Price', 'Terraced House Price', 'Flats Price']

plt.figure(figsize=(12, 6))
for property_type in property_types:
    sns.lineplot(data=medway_df, x='Date', y=property_type, label=property_type)

plt.title('Average Prices of Different Property Types in Medway Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Finding the maximum value in the 'Average price All property types' column
max_average_price = data['Average price All property types'].max()

# Getting the row corresponding to the maximum average price
max_price_row = data[data['Average price All property types'] == max_average_price]

# Displaying the result
print("The highest average price for all property types:")
print(max_price_row)
```

```python
# Finding the minimum value in the 'Average price All property types' column
min_average_price = data['Average price All property types'].min()

# Getting the row corresponding to the maximum average price
min_price_row = data[data['Average price All property types'] == min_average_price]

# Displaying the result
print("The lowest average price for all property types:")
print(min_price_row)
```

```python
# Calculating the mean average price for each county
county_mean_prices = data.groupby('County Name')['Average price All property types'].mean().reset_index()

# Identifying  the top 10 counties with the highest mean average prices
top_10_counties = county_mean_prices.nlargest(10, 'Average price All property types')

# Filtering the original dataset to include only the top 10 counties
top_10_data = data[data['County Name'].isin(top_10_counties['County Name'])]

# Plotting a line graph comparing the average prices over time for the top 10 counties
plt.figure(figsize=(14, 8))

# Plotting each county's data
for county in top_10_data['County Name'].unique():
    county_data = top_10_data[top_10_data['County Name'] == county]
    plt.plot(county_data['Date'], county_data['Average price All property types'], label=county)

plt.xlabel('Date')
plt.ylabel('Average Price (All Property Types)')
plt.title('Comparison of Average House Prices Over Time for Top 10 Counties')
plt.legend(title='County Name')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Calculating the mean average price for each county
county_mean_prices = data.groupby('County Name')['Average price All property types'].mean().reset_index()

# Identifying the bottom 10 counties with the lowest mean average prices
bottom_10_counties = county_mean_prices.nsmallest(10, 'Average price All property types')

# Filtering the original dataset to include only the bottom 10 counties
bottom_10_data = data[data['County Name'].isin(bottom_10_counties['County Name'])]

# Plotting a line graph comparing the average prices over time for the bottom 10 counties
plt.figure(figsize=(14, 8))

# Plotting each county's data
for county in bottom_10_data['County Name'].unique():
    county_data = bottom_10_data[bottom_10_data['County Name'] == county]
    plt.plot(county_data['Date'], county_data['Average price All property types'], label=county)

plt.xlabel('Date')
plt.ylabel('Average Price (All Property Types)')
plt.title('Comparison of Average House Prices Over Time for Bottom 10 Counties')
plt.legend(title='County Name')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Calculating the mean sales volume for each county
county_mean_sales_volumes = data.groupby('County Name')['Sales volume'].mean().reset_index()

# Identifying the top 10 counties with the highest mean sales volumes
top_10_counties_sales_volume = county_mean_sales_volumes.nlargest(10, 'Sales volume')

# Filtering the original dataset to include only the top 10 counties by sales volume
top_10_sales_volume_data = data[data['County Name'].isin(top_10_counties_sales_volume['County Name'])]

# Plot a line graph comparing the sales volumes over time for the top 10 counties
plt.figure(figsize=(14, 8))

# Plot each county's data
for county in top_10_sales_volume_data['County Name'].unique():
    county_data = top_10_sales_volume_data[top_10_sales_volume_data['County Name'] == county]
    plt.plot(county_data['Date'], county_data['Sales volume'], label=county)

plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.title('Comparison of Sales Volumes Over Time for Top 10 Counties')
plt.legend(title='County Name')
plt.xticks(rotation=45)
plt.show()
```

```python
mean_avg_price = data.groupby('County Name')['Average price All property types'].mean().reset_index()

#Selecting the top 10 counties based on the mean average price
top_10_counties = mean_avg_price.nlargest(10, 'Average price All property types')

#Filtering the dataset to include only data from these top 10 counties
top_10_data = data[data['County Name'].isin(top_10_counties['County Name'])]

#Creating the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales volume', y='Average price All property types', hue='County Name', data=top_10_data, palette='tab20', alpha=0.7)
plt.xlabel('Sales Volume')
plt.ylabel('Average Price (All Property Types)')
plt.title('Scatter Plot: Sales Volume vs Average Price for Top 10 Counties')
plt.legend(title='County Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
