This part just combines the Dataset of England, Wales and Scotland seperatley into one CSV file designed for modelling purpose

```python
import pandas as pd

# Lists of CSV file paths that were downloaded seperately
csv_files = ['England_Houses.csv', 'Scotland_Houses.csv', 'Wales_Houses.csv']

# Reading and concatenate the CSV files
combined_csv = pd.concat([pd.read_csv(f) for f in csv_files])

# Saving the concatenated data to a new CSV file
combined_csv.to_csv('combined.csv', index=False)
```
